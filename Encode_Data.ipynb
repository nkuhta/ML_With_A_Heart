```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
train_values = pd.read_csv('train_values.csv')
train_labels = pd.read_csv('train_labels.csv')
test_values = pd.read_csv('test_values.csv')
```

```python
train_values.head()
```

```python
X = train_values.drop(columns='patient_id')
X_test = test_values.drop(columns='patient_id')
X.head()
X_test.head()
```

```python
train_labels.head()
```

```python
y = train_labels.drop(columns='patient_id')
y=np.array(y).flatten()
y
```

#  Create Cross-Validation split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_CV, y_train, y_CV = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
X_train.shape, X_CV.shape
```

#  Encode with pandas get_dummies method
---

```python
# Get dummies
X_train = pd.get_dummies(X_train, prefix_sep='_', drop_first=True)
X_test = pd.get_dummies(X_test, prefix_sep='_', drop_first=True)
X_CV = pd.get_dummies(X_CV, prefix_sep='_', drop_first=True)
```

```python
X_train.head()
X_test.head()
```

```python
X_train.dtypes
```

#  Remove Constant Features

```python
# remove constant features
constant_features = [
    feat for feat in X_train.columns if X_train[feat].std() == 0
]

X_train.drop(labels=constant_features, axis=1, inplace=True)
X_CV.drop(labels=constant_features, axis=1, inplace=True)

X_train.shape, X_CV.shape
```

# Remove Quasi-constant features

```python
from sklearn.feature_selection import VarianceThreshold

# remove quasi-constant features
sel = VarianceThreshold(
    threshold=0.01)  # 0.1 indicates 99% of observations approximately

sel.fit(X_train)  # fit finds the features with low variance

sum(sel.get_support()) # how many not quasi-constant?
```

```python
features_to_keep = X_train.columns[sel.get_support()]
features_to_keep
```

```python
# we can then remove the features like this
X_train = sel.transform(X_train)
X_CV = sel.transform(X_CV)

X_train.shape, X_CV.shape
```

# Duplicate Features

```python
# sklearn transformations lead to numpy arrays
# here I transform the arrays back to dataframes
# please be mindful of getting the columns assigned
# correctly

X_train= pd.DataFrame(X_train)
X_train.columns = features_to_keep

X_CV= pd.DataFrame(X_CV)
X_CV.columns = features_to_keep
```

```python
# check for duplicated features in the training set
duplicated_feat = []
for i in range(0, len(X_train.columns)):
    if i % 10 == 0:  # this helps me understand how the loop is going
        print(i)

    col_1 = X_train.columns[i]

    for col_2 in X_train.columns[i + 1:]:
        if X_train[col_1].equals(X_train[col_2]):
            duplicated_feat.append(col_2)
            
len(duplicated_feat)
```

```python
X_train.drop(labels=duplicated_feat, axis=1, inplace=True)
X_CV.drop(labels=duplicated_feat, axis=1, inplace=True)

X_train.shape, X_CV.shape
```

#  Feature Importance

```python
# Feature Importance with Extra Trees Classifier
from sklearn.ensemble import ExtraTreesClassifier
# load data
# feature extraction
model = ExtraTreesClassifier()
model.fit(X_train, y_train)
print(model.feature_importances_)
```

```python
X_train.columns
```

```python
importance_df = pd.DataFrame()
importance_df['Feature']=X_train.columns
importance_df['Importance']=model.feature_importances_
importance_df.sort_values(by='Importance',ascending=False)
```
